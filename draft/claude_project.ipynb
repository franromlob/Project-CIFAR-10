```python
# =====================================================
# PROJECT I: IMAGE CLASSIFICATION WITH CNN
# Dataset: CIFAR-10
# Optimized for: MacBook M3 (Apple Silicon)
# =====================================================

# 1. IMPORT LIBRARIES
# =====================================================

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report, precision_score, recall_score, f1_score
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping
import warnings
warnings.filterwarnings('ignore')

print(f"TensorFlow version: {tf.__version__}")

# ── M3: Activar Metal GPU ─────────────────────────────────────────────────────
gpus = tf.config.list_physical_devices('GPU')
if gpus:
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
        print(f"✅ Metal GPU detectada: {len(gpus)} dispositivo(s) — memory growth activado")
    except RuntimeError as e:
        print(f"⚠️  Error GPU config: {e}")
else:
    print("ℹ️  Sin GPU — ejecutando en CPU")

print(f"Dispositivos: {[d.device_type for d in tf.config.list_physical_devices()]}")
# ─────────────────────────────────────────────────────────────────────────────

# =====================================================
# 2. LOAD AND EXPLORE CIFAR-10 DATASET
# =====================================================

print("\nDownloading CIFAR-10 dataset...")
(x_train, y_train), (x_test, y_test) = cifar10.load_data()

print(f"\nTraining data shape: {x_train.shape}")
print(f"Test data shape: {x_test.shape}")
print(f"Pixel value range: [{x_train.min()}, {x_train.max()}]")

classes = ['airplane', 'automobile', 'bird', 'cat', 'deer',
           'dog', 'frog', 'horse', 'ship', 'truck']
print(f"Dataset classes: {classes}")

# =====================================================
# 3. VISUALIZE SAMPLE IMAGES
# =====================================================

print("\nDisplaying 10 random images from the dataset...")
fig, axes = plt.subplots(2, 5, figsize=(12, 5))
axes = axes.ravel()

for i in range(10):
    idx = np.random.randint(0, len(x_train))
    axes[i].imshow(x_train[idx])
    axes[i].set_title(f"Class: {classes[y_train[idx][0]]}")
    axes[i].axis('off')

plt.tight_layout()
plt.savefig('sample_images.png', dpi=100, bbox_inches='tight')
plt.show()

# =====================================================
# 4. CLASS DISTRIBUTION
# =====================================================

unique, counts = np.unique(y_train, return_counts=True)
plt.figure(figsize=(10, 5))
plt.bar(classes, counts)
plt.xlabel('Class'); plt.ylabel('Number of Images')
plt.title('Class Distribution in CIFAR-10 (Training Set)')
plt.xticks(rotation=45); plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('class_distribution.png', dpi=100, bbox_inches='tight')
plt.show()

# =====================================================
# 5. DATA PREPROCESSING
# =====================================================

print("\nPreprocessing data...")

x_train_normalized = x_train.astype('float32') / 255.0
x_test_normalized  = x_test.astype('float32')  / 255.0

print(f"Pixel value range after normalization: [{x_train_normalized.min()}, {x_train_normalized.max()}]")

y_train_encoded = to_categorical(y_train, 10)
y_test_encoded  = to_categorical(y_test,  10)

print(f"Shape of y_train after one-hot encoding: {y_train_encoded.shape}")

# =====================================================
# 6. CREATE CNN MODEL
# =====================================================

print("\nCreating CNN model...")

def create_cnn_model():
    model = models.Sequential([
        layers.Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=(32, 32, 3)),
        layers.Conv2D(32, (3, 3), activation='relu', padding='same'),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),

        layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
        layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),

        layers.Flatten(),
        layers.Dense(512, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(10, activation='softmax')
    ])
    return model

model_cnn = create_cnn_model()
model_cnn.summary()

# =====================================================
# 7. COMPILE MODEL
# =====================================================

print("\nCompiling model...")
model_cnn.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# =====================================================
# 8. DATA AUGMENTATION  (tf.data — optimizado para M3)
# =====================================================

print("\nConfiguring data augmentation...")

# ── M3: tf.data + capas de augmentation nativas → mejor rendimiento en Metal
# ── Ajusta BATCH_SIZE según tu RAM unificada:
# ──   8 GB  → 32-64  |  16 GB → 64-128  |  36 GB → 128-256
BATCH_SIZE = 64

data_augmentation = keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.2),
    layers.RandomTranslation(0.1, 0.1),
], name="augmentation")

train_ds_cnn = (
    tf.data.Dataset.from_tensor_slices((x_train_normalized, y_train_encoded))
    .shuffle(10000)
    .batch(BATCH_SIZE)
    .map(lambda x, y: (data_augmentation(x, training=True), y),
         num_parallel_calls=tf.data.AUTOTUNE)
    .prefetch(tf.data.AUTOTUNE)
)

val_ds_cnn = (
    tf.data.Dataset.from_tensor_slices((x_test_normalized, y_test_encoded))
    .batch(BATCH_SIZE)
    .prefetch(tf.data.AUTOTUNE)
)

# =====================================================
# 9. TRAIN CNN MODEL
# =====================================================

print("\nTraining CNN Custom model...")
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

history_cnn = model_cnn.fit(
    train_ds_cnn,
    epochs=100,
    validation_data=val_ds_cnn,
    callbacks=[early_stop],
    verbose=1
)

# =====================================================
# 10. VISUALIZE TRAINING
# =====================================================

fig, axes = plt.subplots(1, 2, figsize=(14, 4))

axes[0].plot(history_cnn.history['accuracy'],     label='Training')
axes[0].plot(history_cnn.history['val_accuracy'], label='Validation')
axes[0].set_xlabel('Epoch'); axes[0].set_ylabel('Accuracy')
axes[0].set_title('Accuracy During Training - CNN Custom')
axes[0].legend(); axes[0].grid(True, alpha=0.3)

axes[1].plot(history_cnn.history['loss'],     label='Training')
axes[1].plot(history_cnn.history['val_loss'], label='Validation')
axes[1].set_xlabel('Epoch'); axes[1].set_ylabel('Loss')
axes[1].set_title('Loss During Training - CNN Custom')
axes[1].legend(); axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('training_metrics_cnn.png', dpi=100, bbox_inches='tight')
plt.show()

# =====================================================
# 11. EVALUATE CNN MODEL
# =====================================================

print("\nEvaluating CNN Custom model...")
test_loss_cnn, test_accuracy_cnn = model_cnn.evaluate(val_ds_cnn, verbose=0)
print(f"\nAccuracy on test set: {test_accuracy_cnn * 100:.2f}%")
print(f"Loss on test set: {test_loss_cnn:.4f}")

y_pred_cnn         = model_cnn.predict(val_ds_cnn)
y_pred_classes_cnn = np.argmax(y_pred_cnn, axis=1)
y_test_classes     = np.argmax(y_test_encoded, axis=1)

# =====================================================
# 12. DETAILED METRICS
# =====================================================

print("\n=== METRICS BY CLASS ===\n")
print(classification_report(y_test_classes, y_pred_classes_cnn, target_names=classes))

precision_cnn = precision_score(y_test_classes, y_pred_classes_cnn, average='weighted')
recall_cnn    = recall_score(y_test_classes,    y_pred_classes_cnn, average='weighted')
f1_cnn        = f1_score(y_test_classes,        y_pred_classes_cnn, average='weighted')

print(f"\n=== GLOBAL METRICS ===")
print(f"Precision: {precision_cnn:.4f}")
print(f"Recall:    {recall_cnn:.4f}")
print(f"F1-Score:  {f1_cnn:.4f}")

# =====================================================
# 13. CONFUSION MATRIX
# =====================================================

cm_cnn = confusion_matrix(y_test_classes, y_pred_classes_cnn)

plt.figure(figsize=(10, 8))
sns.heatmap(cm_cnn, annot=True, fmt='d', cmap='Blues',
            xticklabels=classes, yticklabels=classes)
plt.xlabel('Prediction'); plt.ylabel('True Label')
plt.title('Confusion Matrix - CNN Custom')
plt.tight_layout()
plt.savefig('confusion_matrix_cnn.png', dpi=100, bbox_inches='tight')
plt.show()

# =====================================================
# 14. TRANSFER LEARNING — pipeline M3 optimizado
# =====================================================

print("\n\n=== TRANSFER LEARNING WITH MULTIPLE MODELS ===\n")

# ── M3: El resize ocurre por batches dentro del pipeline — nunca carga
# ──     50.000 imágenes a 224×224 en memoria de golpe (evita el OOM original)
# ──   8 GB RAM → TL_BATCH_SIZE = 16-32
# ──  16 GB RAM → TL_BATCH_SIZE = 32-64
TL_BATCH_SIZE = 32

def make_tl_dataset(images, labels, augment=False):
    """Pipeline tf.data que redimensiona a 224x224 por batch."""
    ds = tf.data.Dataset.from_tensor_slices((images, labels))
    if augment:
        ds = ds.shuffle(10000)
    ds = ds.batch(TL_BATCH_SIZE)
    def preprocess(x, y):
        x = tf.image.resize(x, (224, 224))
        if augment:
            x = data_augmentation(x, training=True)
        return x, y
    return ds.map(preprocess, num_parallel_calls=tf.data.AUTOTUNE).prefetch(tf.data.AUTOTUNE)

train_ds_tl = make_tl_dataset(x_train_normalized, y_train_encoded, augment=True)
val_ds_tl   = make_tl_dataset(x_test_normalized,  y_test_encoded,  augment=False)

early_stop_tl = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# =====================================================
# 14.1 VGG16
# =====================================================

print("1️⃣  TRAINING VGG16...\n")

base_model_vgg = keras.applications.VGG16(
    input_shape=(224, 224, 3), include_top=False, weights='imagenet')
base_model_vgg.trainable = False

model_vgg = models.Sequential([
    base_model_vgg,
    layers.Flatten(),
    layers.Dense(256, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(10, activation='softmax')
])
model_vgg.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    loss='categorical_crossentropy', metrics=['accuracy'])

history_vgg = model_vgg.fit(
    train_ds_tl, epochs=50,
    validation_data=val_ds_tl,
    callbacks=[early_stop_tl], verbose=1)

test_loss_vgg, test_accuracy_vgg = model_vgg.evaluate(val_ds_tl, verbose=0)
print(f"\n✅ VGG16 Accuracy: {test_accuracy_vgg * 100:.2f}%\n")

# =====================================================
# 14.2 MobileNetV2
# =====================================================

print("2️⃣  TRAINING MobileNetV2...\n")

base_model_mobilenet = keras.applications.MobileNetV2(
    input_shape=(224, 224, 3), include_top=False, weights='imagenet')
base_model_mobilenet.trainable = False

model_mobilenet = models.Sequential([
    base_model_mobilenet,
    layers.Flatten(),
    layers.Dense(256, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(10, activation='softmax')
])
model_mobilenet.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    loss='categorical_crossentropy', metrics=['accuracy'])

history_mobilenet = model_mobilenet.fit(
    train_ds_tl, epochs=50,
    validation_data=val_ds_tl,
    callbacks=[early_stop_tl], verbose=1)

test_loss_mobilenet, test_accuracy_mobilenet = model_mobilenet.evaluate(val_ds_tl, verbose=0)
print(f"\n✅ MobileNetV2 Accuracy: {test_accuracy_mobilenet * 100:.2f}%\n")

# =====================================================
# 14.3 InceptionV3
# =====================================================

print("3️⃣  TRAINING InceptionV3...\n")

base_model_inception = keras.applications.InceptionV3(
    input_shape=(224, 224, 3), include_top=False, weights='imagenet')
base_model_inception.trainable = False

model_inception = models.Sequential([
    base_model_inception,
    layers.Flatten(),
    layers.Dense(256, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(10, activation='softmax')
])
model_inception.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    loss='categorical_crossentropy', metrics=['accuracy'])

history_inception = model_inception.fit(
    train_ds_tl, epochs=50,
    validation_data=val_ds_tl,
    callbacks=[early_stop_tl], verbose=1)

test_loss_inception, test_accuracy_inception = model_inception.evaluate(val_ds_tl, verbose=0)
print(f"\n✅ InceptionV3 Accuracy: {test_accuracy_inception * 100:.2f}%\n")

# =====================================================
# 15. COMPARISON OF ALL MODELS
# =====================================================

print("\n" + "="*50)
print("=== COMPARISON OF ALL MODELS ===")
print("="*50 + "\n")

results = {
    'CNN Custom':  test_accuracy_cnn       * 100,
    'VGG16':       test_accuracy_vgg       * 100,
    'MobileNetV2': test_accuracy_mobilenet * 100,
    'InceptionV3': test_accuracy_inception * 100
}

for model_name, accuracy in sorted(results.items(), key=lambda x: x[1], reverse=True):
    print(f"{model_name:.<30} {accuracy:.2f}%")

best_model_name = max(results, key=results.get)
best_accuracy   = results[best_model_name]
print(f"\n🏆 BEST MODEL: {best_model_name} ({best_accuracy:.2f}%)")

fig, ax = plt.subplots(figsize=(10, 6))
model_names = list(results.keys())
accuracies  = list(results.values())
colors = ['gold' if m == best_model_name else 'steelblue' for m in model_names]

bars = ax.bar(model_names, accuracies, color=colors, edgecolor='black', linewidth=1.5)
ax.set_ylabel('Accuracy (%)', fontsize=12)
ax.set_title('Accuracy Comparison - All Models', fontsize=14, fontweight='bold')
ax.set_ylim([0, 100]); ax.grid(axis='y', alpha=0.3)

for bar, accuracy in zip(bars, accuracies):
    ax.text(bar.get_x() + bar.get_width()/2., bar.get_height(),
            f'{accuracy:.2f}%', ha='center', va='bottom', fontweight='bold')

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('models_comparison.png', dpi=100, bbox_inches='tight')
plt.show()

# =====================================================
# 16. COMPARE TRAINING CURVES
# =====================================================

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

for ax, history, title in zip(
    axes.ravel(),
    [history_cnn, history_vgg, history_mobilenet, history_inception],
    ['CNN Custom', 'VGG16', 'MobileNetV2', 'InceptionV3']
):
    ax.plot(history.history['accuracy'],     label='Training')
    ax.plot(history.history['val_accuracy'], label='Validation')
    ax.set_title(title, fontsize=12, fontweight='bold')
    ax.set_ylabel('Accuracy'); ax.legend(); ax.grid(True, alpha=0.3)

axes[1, 0].set_xlabel('Epoch')
axes[1, 1].set_xlabel('Epoch')
plt.tight_layout()
plt.savefig('training_curves_comparison.png', dpi=100, bbox_inches='tight')
plt.show()

# =====================================================
# 17. BEST MODEL JUSTIFICATION
# =====================================================

print("\n" + "="*50)
print("=== BEST MODEL JUSTIFICATION ===")
print("="*50 + "\n")

justifications = {
    'CNN Custom': """
    - Lightweight and fast to train
    - Low memory usage
    - Good for learning deep learning basics
    - Does not leverage pre-trained features
    """,
    'VGG16': """
    - Simple and clear architecture
    - Good classification accuracy
    - More complex than CNN Custom
    - Requires more training time
    """,
    'MobileNetV2': """
    - Optimized for mobile devices
    - Very efficient in resource usage
    - Good accuracy-to-speed ratio
    - Ideal for deployment on resource-limited devices
    """,
    'InceptionV3': """
    - Complex and powerful architecture
    - Excellent at capturing complex features
    - High accuracy but requires more resources
    - Best for large and complex datasets
    """
}

print(f"✅ SELECTED BEST MODEL: {best_model_name}\n")
print(justifications[best_model_name])

# =====================================================
# 18. SAVE ALL MODELS
# =====================================================

print("\nSaving all models...\n")

# ── M3: formato .keras recomendado sobre .h5 para TF >= 2.12 ─────────────────
model_cnn.save('model_cnn_custom_cifar10.keras');        print("✅ CNN Custom saved")
model_vgg.save('model_vgg16_cifar10.keras');             print("✅ VGG16 saved")
model_mobilenet.save('model_mobilenetv2_cifar10.keras'); print("✅ MobileNetV2 saved")
model_inception.save('model_inceptionv3_cifar10.keras'); print("✅ InceptionV3 saved")

best_model_map = {
    'CNN Custom':  model_cnn,
    'VGG16':       model_vgg,
    'MobileNetV2': model_mobilenet,
    'InceptionV3': model_inception
}
best_model = best_model_map[best_model_name]
best_model.save('best_model_cifar10.keras')
print(f"\n🏆 Best model ({best_model_name}) saved as 'best_model_cifar10.keras'")

# =====================================================
# 19. PREDICTIONS WITH BEST MODEL
# =====================================================

print("\n" + "="*50)
print("=== PREDICTIONS WITH BEST MODEL ===")
print("="*50 + "\n")

def predict_single(image_normalized, model, is_transfer=False):
    """Predice una sola imagen normalizada 32x32."""
    img = np.expand_dims(image_normalized, axis=0)   # (1,32,32,3)
    if is_transfer:
        img = tf.image.resize(img, (224, 224))        # (1,224,224,3)
    return model.predict(img, verbose=0)[0]

is_transfer = (best_model_name != 'CNN Custom')
print(f"Predictions with {best_model_name}:\n")

for i in range(5):
    idx   = np.random.randint(0, len(x_test))
    probs = predict_single(x_test_normalized[idx], best_model, is_transfer)

    predicted_class = classes[np.argmax(probs)]
    confidence      = np.max(probs) * 100
    true_class      = classes[y_test_classes[idx]]
    is_correct      = "✅" if predicted_class == true_class else "❌"

    print(f"Sample {i+1}: {is_correct}")
    print(f"  True class:  {true_class}")
    print(f"  Prediction:  {predicted_class}")
    print(f"  Confidence:  {confidence:.2f}%\n")

print("\n✅ Project completed successfully!")
```
